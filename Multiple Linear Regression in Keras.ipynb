케라스는 딥러닝 모델을 만들기 위한 고수준의 구성 요소를 제공하는 모델 수준의 라이브러리입니다.

```python
# TensorFlow and tf.keras
import tensorflow as tf
import tensorflow.keras.models as models
import tensorflow.keras.layers as layers

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

import sys
print('Python Ver. : ', sys.version)
print('Tensorflow Ver. : ', tf.__version__)
```

일차 함수 (y = ax1 + bx2 + cx3 + d)를 DL의 Neural Net으로 어떻게 모델링 할 수 있는지 테스트해 본다

```python
# y = ax1 + bx2 + cx3 + d의 데이터 (a=3, b=1, c= 2, d=0.3)
x_data = np.array([[1, 2, 3], 
                   [2, 0, 2],
                   [4, 3, -1],
                   [7, 5, -3],
                   [10, 2, -7]])
```

```python
y_data = np.dot(x_data, [[3],[1],[2]])+0.3
```

```python
y_data
```

## 모델 구성

Keras로 모델을 만드는 세 가지 방법이 있다.  
1. Sequential API
2. Functional API
3. Model Subclassing

```python
# Sequential API
model = models.Sequential()
model.add(layers.Dense(1, activation='linear', input_shape=(3,)))

# 다음과 같이 묶는 방법도 있다
#model = tf.keras.Sequential([
#    tf.keras.layers.Dense(1, input_dim=3, activation='linear')
#])
```

```python
# Functional API를 사용하는 방법은 다음과 같다
# Functional API
#input_tensor = layers.Input(shape=(3,))
#output_tensor = layers.Dense(1, activation='linear')(input_tensor)
#model = models.Model(inputs=input_tensor, outputs=output_tensor)
```

```python
# 모델의 구조 살펴 보기
model.summary()
```

## 모델 컴파일

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.01),
              loss='mse',
              metrics=['mse'])
```

## 모델 훈련

```python
model.fit(x_data, y_data, batch_size=1, epochs=2000, shuffle=False)
```

## 예측하기

```python
x =np.array([[11,2,-3]])
print(model.predict(x))
```

## 가중치 확인하기
